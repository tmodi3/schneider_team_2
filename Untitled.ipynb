```python
import csv
import time

def load_csv(file_path):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)  
        data = [row for row in reader]  
    return data

def test_helper(idpost, stop_data):
    stops = [row for row in stop_data if row['LOAD_ID'] == idpost]
    return stops

def testing(type_truck, radius, weight, posting_data, stop_data, origin, dest):
    start_time = time.time()
    
    result = [row for row in posting_data if float(row['TOTAL_WEIGHT']) <= weight 
              and float(row['TOTAL_DISTANCE']) <= radius
              and row['TRANSPORT_MODE'] == type_truck]
    
    ids= [row["LOAD_ID"] for row in result]
    
    matching_origin = []
    
    for i in ids:
        stops = test_helper(i, stop_data)
        ori = False
        des = False
        
        for stop in stops:
            if stop["STOP_SEQUENCE"] == "1":
                if stop["STATE"] == origin: 
                    ori = True
                    
            if stop["STOP_SEQUENCE"] == str(len(stops)):
                if stop["STATE"] == dest:
                    des = True
        
            if ori and des:
                
                matching_origin.append(i)
            
            
    print(matching_origin)
    
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    
    print(f"Test Case 1 Result:")
    
    
    print(f"Time taken: {elapsed_time:.6f} seconds\n")
    return elapsed_time


```

```python
def random_testing(posting_data, stop_data, num_tests=100):
    truck_types = ['Power Only', 'Dry Van', 'Refrigerated', 'Power Only Team']
    states = ['GA', 'OK', 'TX', 'MO', 'KS', 'OH', 'CO', 'VA', 'IA', 'PA', 'MD',
       'NC', 'TN', 'KY', 'IL', 'CA', 'IN', 'MI', 'SC', 'NY', 'UT', 'FL',
       'WI', 'NJ', 'MN', 'OR', 'NH', 'NV', 'AR', 'MS', 'AZ', 'LA', 'CT',
       'NE', 'AL', 'MA', 'WA', 'WV', 'DE', 'ID', 'NM', 'ON', 'ME', 'ND',
       'WY', 'RI', 'VT']
    total_times = []
    
    for i in range(num_tests):
        type_truck = random.choice(truck_types)
        radius = random.uniform(50, 500)  
        weight = random.uniform(35000, 50000)
        origin = random.choice(states)
        dest = origin
        
        print(f"Test {i+1}: type_truck={type_truck}, radius={radius:.2f}, weight={weight:.2f}, origin={origin}, dest={dest}")
        
        time_taken = testing(type_truck, radius, weight, posting_data, stop_data, origin, dest)
        total_times.append(time_taken)
    
    avg_time = sum(total_times) / num_tests
    print(f"Average time taken for {num_tests} searches: {avg_time:.6f} seconds")
    print(f"Total time taken: {sum(total_times)}")

    

    
```
